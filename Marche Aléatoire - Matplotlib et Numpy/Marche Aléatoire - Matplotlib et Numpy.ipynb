EXERCICE 1 - Vous implémenterez cette marche pour différentes valeurs de p et N et vous calculerez pour chaque expérience le déplacement obtenu après N pas.

```python
import numpy
```

```python
p = numpy.random.rand()    # N'importe quelle valeur pour p (probabilité d'aller de l'avant)
N = numpy.random.randint(1,10)  # Une valeur entière pour N entre 1 et 10 (peut être changée en valeurs plus grandes si désiré)
q = 1-p # probabilité de retour en arrière
print(p)
print(N)
```

```python
random_number = numpy.random.rand(N) # Valeurs aléatoires entre 0 et 1, N fois
print(random_number)
```

```python
# Le deplacement est le nombre de pas effectués en avant (+1) moins le nombre de pas effectués en arrière (-1)
deplacement = int((numpy.size(random_number[random_number > q])) - numpy.size(random_number[random_number <= q]))
print(f'Le deplacement est: {deplacement}')
```

```python
# Même opération, mais ici nous ajoutons la moyenne et l'écart-type des valeurs pour M=10, M=100, M=1000 et M=10000 dans deux vecteurs séparés.

deplacementTotal = []
moyen = []
ecart_types = []
M = 10
MC = 0

while M < 10001:
    while MC < M:
        MC = MC + 1
        random_number = numpy.random.rand(N)
        deplacement = int((numpy.size(random_number[random_number > q])) - numpy.size(random_number[random_number <= q]))
        deplacementTotal.append(deplacement)
    moyen.append(numpy.mean(deplacementTotal))
    ecart_types.append(numpy.std(deplacementTotal))
    M = M*10
```

```python
# Matrice 2x4 avec moyennes et écart-type
tableau = numpy.array([moyen, ecart_types])
print(f'moyen: {moyen}')
print(f'ecart_types: {ecart_types}')
print(f'tableau: \n{tableau}')
```

```python
# Marche 1D
def marcher(q, N):
    random_number = numpy.random.random(N)
    random_number[random_number > q] = 1
    random_number[random_number <= q] = -1
    return numpy.concatenate((numpy.array([0]),numpy.cumsum(random_number)),axis=None)
```

EXERCICE 2 - Généralisez la marche aléatoire 1D au cas 2D et représentez son déplacement. Vous choisirez la représentation qui vous paraît être la plus appropriée (couleur, titre, légendes doivent être telles que la courbe doit se suffise à elle-même).

Attention le déplacement doit se faire perpendiculairement aux axes x et y.

```python
# Marche 2D

def marcher2D(qH, qV, N):
    random_number_horizontal = numpy.random.random(N)
    random_number_vertical = numpy.random.random(N)
    
    random_number_horizontal[random_number_horizontal > qH] = 1
    random_number_vertical[random_number_vertical > qV] = 1

    random_number_horizontal[random_number_horizontal <= qH] = -1
    random_number_vertical[random_number_vertical <= qV] = -1

    #R épéter les valeurs pour ne pas bouger sur les 2 axes en même temps
    horizontal_rep = numpy.repeat(numpy.cumsum( random_number_horizontal), 2)
    vertical_rep = numpy.repeat(numpy.cumsum(random_number_vertical), 2)

    return numpy.concatenate((numpy.array([0]), horizontal_rep),axis=None),numpy.concatenate((numpy.array([0,0]),vertical_rep),axis=None)
```

```python
# Valeurs de probabilité aléatoires pour le déplacement horizontal
pH = numpy.random.rand()
qH = 1 - pH

# Valeurs de probabilité aléatoires pour le déplacement vertical
pV = numpy.random.rand()
qV = 1 - pV

marcher = marcher2D(qH,qV,N)
```

```python
# Répéter la dernière valeur du tableau X afin que les deux tableaux (X et Y) aient le même nombre de coordonnées
deplacement_x = numpy.append(marcher[0], marcher[0][-1])

# Mettre des informations sur le graphique
matplotlib.pyplot.plot(deplacement_x, marcher[1],".--",linewidth=1, color ='blue')
```
